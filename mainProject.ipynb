```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import metrics
```

```python
dataset=pd.read_csv('D:\downloads\TSLA.csv')
```

```python
print(dataset)
```

```python
x=dataset[['Open','High','Low','Volume']]
y=dataset[['Close']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=.25,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
rg=LinearRegression()
```

```python
rg.fit(x_train,y_train)
```

```python
print(rg.coef_)
```

```python
predicted=rg.predict(x_test)
```

```python
predicted.shape
```

```python
y_test.shape
```

```python
print(predicted)
```

```python
pred= predicted.reshape(604)
```

```python
print(y_test)
```

```python
dframe=pd.DataFrame(y_test['Close'],pred)
```

```python
dout=pd.DataFrame({'Actual price:':y_test['Close'],'Predicted price:':pred})
```

```python
print(dout)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print("Accuracy score:",rg.score(x_test,y_test))
```

```python
print("Mean absolute error:",metrics.mean_absolute_error(y_test['Close'],pred))
```

```python
print("Mean squared error:",metrics.mean_squared_error(y_test['Close'],pred))
```
